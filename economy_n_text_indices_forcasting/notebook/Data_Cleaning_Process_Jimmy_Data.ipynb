```python
import pandas as pd
import numpy as np

RAW_CSV = "Jimmy_data_renewed.csv"

PROGRAMS_TO_FILTER = [
    'Tax Clinic',
    'Newcomer Community Integration Program (NCIP)',
    'Financial Empowerment Counselling',
    'Integrated Employment Services',
    'Family Newcomer Services',
    'Financial Empowerment Library Project',
    'Tax Skills for Self-Employed Workers',
    'Language Instruction for Newcomers to Canada',
    'Visual Effects Compositing Program',
    'Older Worker Initiative',
    'Employment Ontario',
    'Project: Income Transition Project',
    'Newcomer Settlement and Integration Services (Newcomer Services Program)',
    'Family Finance Clinic',
    'Asylum Integration Program (AIP)',
    'Community Connections - Group Activities',
    'Employment Accessibility Services',
    'Personalized Career Accelerator (PCA)',
    'Youth Housing Navigation and Subsidy Support Program',
    'Newcomer Women Entrepreneurship Program',
    'Community Connections - Mentorship',
    'Enhanced Language Training',
    'Newcomer Youth Career Start (NYCS)',
    'Youth Job Connections (YJC)',
    'CNC/Precision Machining Skills Training Program',
    'Plumbing Pre-Apprenticeship Level 1 Program',
    'Toronto Youth Job Corp',
    'Youth Job Connections Summer (YJC Summer)',
    'Industrial Millwright Pre-Apprenticeship Program'
]

WEEKLY_RULE = "W-SUN"


def build_weekly_program_intake(
    raw_csv: str,
    programs: list[str],
    weekly_rule: str = "W-SUN",
) -> pd.DataFrame:
    df = pd.read_csv(raw_csv, low_memory=False)

    df["program_name_DC"] = df["program_name_DC"].fillna("General_purposes")
    df = df.dropna(subset=["Creation Date"])

    df["Creation Date"] = pd.to_datetime(df["Creation Date"], format="mixed")
    # df = df.iloc[2:] # Drop the first 2 rows
    df = df[df["program_name_DC"].isin(programs)]

    daily = (
        df.groupby(["Creation Date", "program_name_DC"])
          .size()
          .reset_index(name="number_of_people")
    )

    pivot = daily.pivot_table(
        index="Creation Date",
        columns="program_name_DC",
        values="number_of_people",
        aggfunc="sum",
    ).fillna(0).sort_index()

    pivot = pivot.iloc[2:]
    weekly = pivot.resample(weekly_rule).sum()
    weekly = weekly.asfreq(weekly_rule, fill_value=0)

    weekly.index.name = "week_end"
    return weekly.reset_index()


def build_weekly_total_intake(
    raw_csv: str,
    programs: list[str],
    weekly_rule: str = "W-SUN",
) -> pd.DataFrame:
    df = pd.read_csv(raw_csv, low_memory=False)

    df["program_name_DC"] = df["program_name_DC"].fillna("General_purposes")
    df = df.dropna(subset=["Creation Date"])

    df["Creation Date"] = pd.to_datetime(df["Creation Date"], format="mixed")
    # df = df.iloc[2:] # Drop the first 2 rows
    # print(df.head(2))
    df = df[df["program_name_DC"].isin(programs)]

    daily = (
        df.groupby("Creation Date")
          .size()
          .to_frame("number_of_people")
          .sort_index()
    )

    daily = daily.iloc[2:]
    weekly = daily.resample(weekly_rule).sum()
    weekly = weekly.asfreq(weekly_rule, fill_value=0)

    weekly.index.name = "week_end"
    return weekly.reset_index()


# === build and save canonical weekly data ===
weekly_program_intake = build_weekly_program_intake(
    raw_csv=RAW_CSV,
    programs=PROGRAMS_TO_FILTER,
    weekly_rule=WEEKLY_RULE,
)
weekly_program_intake.to_csv("program_intake_weekly.csv", index=False)

weekly_total_intake = build_weekly_total_intake(
    raw_csv=RAW_CSV,
    programs=PROGRAMS_TO_FILTER,
    weekly_rule=WEEKLY_RULE,
)
weekly_total_intake.to_csv("total_intake_weekly.csv", index=False)
```

```python
import pandas as pd

# Assuming RAW_CSV and PROGRAMS_TO_FILTER are already defined in the notebook
# RAW_CSV = 'Jimmy_data_renewed.csv'
# PROGRAMS_TO_FILTER = [...]

def build_daily_program_intake(
    raw_csv: str,
    programs: list[str],
) -> pd.DataFrame:
    df = pd.read_csv(raw_csv, low_memory=False)

    df["program_name_DC"] = df["program_name_DC"].fillna("General_purposes")
    df = df.dropna(subset=["Creation Date"])

    df["Creation Date"] = pd.to_datetime(df["Creation Date"], format="mixed")
    df = df[df["program_name_DC"].isin(programs)]

    daily = (
        df.groupby(["Creation Date", "program_name_DC"])
          .size()
          .reset_index(name="number_of_people")
    )

    pivot = daily.pivot_table(
        index="Creation Date",
        columns="program_name_DC",
        values="number_of_people",
        aggfunc="sum",
    ).fillna(0).sort_index()

    # Resample to daily frequency
    daily_agg = pivot.resample('D').sum()
    daily_agg = daily_agg.asfreq('D', fill_value=0)

    daily_agg.index.name = "date"
    return daily_agg.reset_index()


def build_daily_total_intake(
    raw_csv: str,
    programs: list[str],
) -> pd.DataFrame:
    df = pd.read_csv(raw_csv, low_memory=False)

    df["program_name_DC"] = df["program_name_DC"].fillna("General_purposes")
    df = df.dropna(subset=["Creation Date"])

    df["Creation Date"] = pd.to_datetime(df["Creation Date"], format="mixed")
    df = df[df["program_name_DC"].isin(programs)]

    daily = (
        df.groupby("Creation Date")
          .size()
          .to_frame("number_of_people")
          .sort_index()
    )

    # Resample to daily frequency
    daily_agg = daily.resample('D').sum()
    daily_agg = daily_agg.asfreq('D', fill_value=0)

    daily_agg.index.name = "date"
    return daily_agg.reset_index()


# === build and save canonical daily data ===
daily_program_intake = build_daily_program_intake(
    raw_csv=RAW_CSV,
    programs=PROGRAMS_TO_FILTER,
)
# Drop the first 2 rows
daily_program_intake = daily_program_intake.iloc[2:]
daily_program_intake.to_csv("program_intake_daily.csv", index=False)
print('Saved program_intake_daily.csv')

daily_total_intake = build_daily_total_intake(
    raw_csv=RAW_CSV,
    programs=PROGRAMS_TO_FILTER,
)
# Drop the first 2 rows
daily_total_intake = daily_total_intake.iloc[2:]
daily_total_intake.to_csv("total_intake_daily.csv", index=False)
print('Saved total_intake_daily.csv')
```
