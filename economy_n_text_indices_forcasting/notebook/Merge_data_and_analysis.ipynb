```python
import pandas as pd

# Load the datasets
df_time_series = pd.read_csv('/content/new_time_series_data.csv')
df_processed = pd.read_csv('/content/df_processed.csv')
df_index_processed = pd.read_csv('/content/df_index_processed.csv')

# Display the first few rows and info of each DataFrame to understand their structure
print("new_time_series_data.csv head:")
display(df_time_series.head())
print("\nnew_time_series_data.csv info:")
display(df_time_series.info())

print("\ndf_processed.csv head:")
display(df_processed.head())
print("\ndf_processed.csv info:")
display(df_processed.info())

print("\ndf_index_processed.csv head:")
display(df_index_processed.head())
print("\ndf_index_processed.csv info:")
display(df_index_processed.info())
```

## Calculate Program Consistency using Mean, CV, Days Active

Calculate the mean, standard deviation, Coefficient of Variation (CV), and the percentage of days a program had an intake (number of people > 0) for each program using the df_time_series_programs DataFrame.


```python
program_data = df_time_series_programs.drop(columns=['date'])

# Calculate the percentage of days each program had an intake (number of people > 0)
program_pct_active_days = {}
for column in program_data.columns:
    total_days = len(program_data)
    active_days = (program_data[column] > 0).sum()
    program_pct_active_days[column] = (active_days / total_days) * 100

program_pct_active_days = pd.Series(program_pct_active_days)

print("Percentage of active days for each program:")
display(program_pct_active_days.sort_values(ascending=False))
```

```python
import numpy as np

# Calculate the mean for each program
program_means = program_data.mean()

# Calculate the standard deviation for each program
program_stds = program_data.std()

# Calculate the Coefficient of Variation (CV) for each program
# Handle potential division by zero by setting CV to NaN if mean is 0
program_cvs = program_stds / program_means
program_cvs = program_cvs.replace([np.inf, -np.inf], np.nan) # Replace inf with NaN

# Create a DataFrame to consolidate all metrics
program_metrics = pd.DataFrame({
    'Mean': program_means,
    'Standard Deviation': program_stds,
    'CV': program_cvs,
    'Pct_Days_Active': program_pct_active_days
})

print("Consolidated Program Metrics:")
display(program_metrics.sort_values(by='CV', ascending=True))
```

## Visualize Program Consistency (CV and Pct_Days_Active)



```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create a bar chart for CV, sorted by CV (ascending)
plt.figure(figsize=(12, 8))
sns.barplot(x=program_metrics['CV'].sort_values(ascending=True).values, y=program_metrics['CV'].sort_values(ascending=True).index, palette='viridis', hue=program_metrics['CV'].sort_values(ascending=True).index, legend=False)
plt.title('Program Consistency: Coefficient of Variation (CV)', fontsize=16)
plt.xlabel('Coefficient of Variation (CV)', fontsize=12)
plt.ylabel('Program', fontsize=12)
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

# Create a bar chart for Pct_Days_Active, sorted by Pct_Days_Active (descending)
plt.figure(figsize=(12, 8))
sns.barplot(x=program_metrics['Pct_Days_Active'].sort_values(ascending=False).values, y=program_metrics['Pct_Days_Active'].sort_values(ascending=False).index, palette='magma', hue=program_metrics['Pct_Days_Active'].sort_values(ascending=False).index, legend=False)
plt.title('Program Activity: Percentage of Days with Intake', fontsize=16)
plt.xlabel('Percentage of Days Active', fontsize=12)
plt.ylabel('Program', fontsize=12)
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Ensure the dataframe is sorted by date for time-series operations
merged_df_cleaned = merged_df_cleaned.sort_values(by='date').reset_index(drop=True)

# --- Lagged Correlations ---
print("\n--- Lagged Correlations ---")

target_variable = 'number of people'
# Select all features for lagged correlation analysis, excluding 'date' and the target itself
all_columns = merged_df_cleaned.columns.tolist()
features_for_lagged_corr = [col for col in all_columns if col not in ['date', target_variable, 'number_of_people_MA7', 'number_of_people_MA30']]

max_lag = 14 # Analyze correlations up to 14 days lag

lagged_correlations = pd.DataFrame(index=features_for_lagged_corr, columns=range(1, max_lag + 1))

for feature in features_for_lagged_corr:
    for lag in range(1, max_lag + 1):
        # Create a lagged version of the feature
        lagged_feature = merged_df_cleaned[feature].shift(lag)
        # Calculate correlation, dropping NaNs introduced by shifting
        correlation = merged_df_cleaned[target_variable].corr(lagged_feature)
        lagged_correlations.loc[feature, lag] = correlation

print("Lagged Correlations with 'number of people':")
display(lagged_correlations)

# Visualize lagged correlations
plt.figure(figsize=(20, 15)) # Increased figure size for more features
sns.heatmap(lagged_correlations.astype(float), annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title(f'Lagged Correlation of All Features with {target_variable}')
plt.xlabel('Lag (Days)')
plt.ylabel('Feature')
plt.show()

# --- Moving Averages ---
print("\n--- Moving Averages ---")

# Calculate moving averages for 'number of people'
merged_df_cleaned['number_of_people_MA7'] = merged_df_cleaned[target_variable].rolling(window=7, min_periods=1).mean()
merged_df_cleaned['number_of_people_MA30'] = merged_df_cleaned[target_variable].rolling(window=30, min_periods=1).mean()

# Plot 'number of people' with its moving averages
plt.figure(figsize=(15, 7))
sns.lineplot(x='date', y=target_variable, data=merged_df_cleaned, label='Original')
sns.lineplot(x='date', y='number_of_people_MA7', data=merged_df_cleaned, label='7-Day Moving Average')
sns.lineplot(x='date', y='number_of_people_MA30', data=merged_df_cleaned, label='30-Day Moving Average')
plt.title(f'{target_variable} with Moving Averages')
plt.xlabel('Date')
plt.ylabel('Number of People')
plt.legend()
plt.grid(True)
plt.show()

print("First few rows of merged_df_cleaned with moving averages:")
display(merged_df_cleaned[[target_variable, 'number_of_people_MA7', 'number_of_people_MA30']].head())
```
