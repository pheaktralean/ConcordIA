```python
import re, numpy as np, pandas as pd, matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
import warnings
warnings.filterwarnings("ignore")
```

```python
dc_cols = [c for c in df.columns if c.endswith("_DC") or c in ["Unique Record ID"]]
df_dc = df[dc_cols].copy()
print(f"Selected {len(dc_cols)} DC columns")
```

```python
df_dc["program_startdate_DC"] = pd.to_datetime(df_dc["program_startdate_DC"], errors="coerce")
df_dc = df_dc.dropna(subset=["program_startdate_DC"])
```

```python
df_dc["employmentstatus_DC"] = (
    df_dc["employmentstatus_DC"]
    .astype(str)
    .str.strip()
    .replace(["", "none", "nan", "n/a", "-", "--"], np.nan, regex=True)
)

print("Top statuses:")
print(df_dc["employmentstatus_DC"].value_counts(dropna=False).head(10))
```

```python
def is_employed(val, work_any=None, looking=None):
    """Return 1 if employed, 0 if not, np.nan if unknown."""
    if isinstance(val, str):
        s = val.lower()
        if any(x in s for x in ["full", "part", "self", "employ", "working"]):
            return 1
        if any(x in s for x in ["unemployed", "not in labor", "retired"]):
            return 0
    if isinstance(work_any, str) and work_any.strip().lower() == "yes":
        return 1
    if isinstance(looking, str) and looking.strip().lower() == "yes":
        return 0
    return np.nan

df_dc["is_employed"] = df_dc.apply(
    lambda r: is_employed(
        r.get("employmentstatus_DC"),
        r.get("mir_workwithanyemployer_DC"),
        r.get("mir_looking_for_work_DC"),
    ),
    axis=1,
)

print("\nEmployment flag summary:")
print(df_dc["is_employed"].value_counts(dropna=False))
```

```python
df_dc["month"] = df_dc["program_startdate_DC"].dt.to_period("M").dt.to_timestamp()
monthly_dc = (
    df_dc.groupby("month")
    .agg(
        total_clients=("Unique Record ID", "nunique"),
        employed_clients=("is_employed", lambda x: (x == 1).sum()),
    )
    .reset_index()
)

monthly_dc["employment_rate"] = monthly_dc["employed_clients"] / monthly_dc["total_clients"]
monthly_dc = monthly_dc.set_index("month").asfreq("MS").fillna(0).reset_index()

print("\nTail of monthly rollup:")
print(monthly_dc.tail())
```

```python
plt.figure(figsize=(10,5))
plt.plot(monthly_dc["month"], monthly_dc["employment_rate"], marker="o")
plt.title("Monthly Employment Rate (DC system)")
plt.xlabel("Month")
plt.ylabel("Employment Rate")
plt.grid(True)
plt.show()
```

```python
ts = monthly_dc.set_index("month")["employment_rate"].asfreq("MS")
if ts.notna().sum() >= 6:
    model = ARIMA(ts, order=(1,1,1))
    fit = model.fit()
    forecast = fit.forecast(steps=3)
    forecast.index = pd.date_range(ts.index[-1] + pd.offsets.MonthBegin(), periods=3, freq="MS")

    plt.figure(figsize=(10,5))
    plt.plot(ts, label="Observed")
    plt.plot(forecast, label="Forecast", color="orange")
    plt.title("Employment Rate Forecast (3 Months Ahead)")
    plt.legend()
    plt.grid(True)
    plt.show()
else:
    print("Not enough months with non-missing data for forecasting.")

monthly_dc.to_csv("employment_cleaned_DC.csv", index=False)
print("Saved cleaned monthly employment data to employment_cleaned_DC.csv")
```

```python
# Your employment programs list
employment_programs = [
    "integrated employment services",
    "employment accessibility services",
    "employment ontario",
    "personalized career accelerator (pca)",
    "older worker initiative",
    "youth job connections (yjc)",
    "youth job connections summer (yjc summer)",
    "newcomer youth career start (nycs)",
    "cnc/precision machining skills training program",
    "plumbing pre-apprenticeship level 1 program",
    "industrial millwright pre-apprenticeship program",
    "newcomer women entrepreneurship program",
    "tax skills for self-employed workers",
    "enhanced language training",
    "visual effects compositing program",
]

# Normalize program names for matching
df['program_normalized'] = df['program_name_DC'].fillna('').str.lower().str.strip()

# Filter for employment programs
employment_df = df[df['program_normalized'].isin(employment_programs)].copy()

print(f"Total employment program registrations: {len(employment_df):,}")
```

```python
# Convert program_startdate_DC to datetime and extract year
employment_df['program_startdate_DC'] = pd.to_datetime(
    employment_df['program_startdate_DC'], 
    errors='coerce'
)
employment_df['start_year'] = employment_df['program_startdate_DC'].dt.year

# Remove rows where year couldn't be extracted
employment_df_with_year = employment_df.dropna(subset=['start_year'])

print(f"Registrations with valid dates: {len(employment_df_with_year):,}")
print(f"Missing dates: {len(employment_df) - len(employment_df_with_year):,}")
```

```python
# Aggregate registrations by year
yearly_data = employment_df_with_year.groupby('start_year').size().reset_index(name='registrations')
yearly_data = yearly_data.sort_values('start_year')

# Convert year to datetime for time series analysis
yearly_data['date'] = pd.to_datetime(yearly_data['start_year'].astype(int).astype(str) + '-01-01')

print("Historical Data:")
print(yearly_data)

# Visualize the trend
plt.figure(figsize=(12, 6))
plt.plot(yearly_data['start_year'], yearly_data['registrations'], marker='o', linewidth=2, markersize=8)
plt.title('Historical Employment Program Registrations', fontsize=16, fontweight='bold')
plt.xlabel('Year', fontsize=12)
plt.ylabel('Number of Registrations', fontsize=12)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np

# Prepare data
X = yearly_data['start_year'].values.reshape(-1, 1)
y = yearly_data['registrations'].values

# Train model
lr_model = LinearRegression()
lr_model.fit(X, y)

# Make predictions
y_pred = lr_model.predict(X)

# Forecast next year
next_year = yearly_data['start_year'].max() + 1
next_year_pred = lr_model.predict([[next_year]])[0]

print("LINEAR REGRESSION FORECAST")
print("="*60)
print(f"Next Year ({int(next_year)}): {int(next_year_pred):,} registrations")
print(f"\nModel Performance:")
print(f"  R Score: {r2_score(y, y_pred):.4f}")
print(f"  MAE: {mean_absolute_error(y, y_pred):.2f}")
print(f"  RMSE: {np.sqrt(mean_squared_error(y, y_pred)):.2f}")
```

```python
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

# Fit ARIMA model (auto-detect best parameters)
# Try different orders
best_aic = np.inf
best_order = None
best_model = None

for p in range(0, 3):
    for d in range(0, 2):
        for q in range(0, 3):
            try:
                model = ARIMA(yearly_data['registrations'], order=(p, d, q))
                fitted_model = model.fit()
                if fitted_model.aic < best_aic:
                    best_aic = fitted_model.aic
                    best_order = (p, d, q)
                    best_model = fitted_model
            except:
                continue

print("ARIMA MODEL FORECAST")
print("="*60)
print(f"Best ARIMA order: {best_order}")
print(f"AIC: {best_aic:.2f}")

# Forecast next year
forecast = best_model.forecast(steps=1)
print(f"\nNext Year ({int(next_year)}): {int(forecast.iloc[0]):,} registrations")

# Show model summary
print("\nModel Summary:")
print(best_model.summary())
```

```python
from prophet import Prophet

# Prepare data for Prophet (needs 'ds' and 'y' columns)
prophet_data = yearly_data[['date', 'registrations']].copy()
prophet_data.columns = ['ds', 'y']

# Create and fit model
prophet_model = Prophet(
    yearly_seasonality=True,
    changepoint_prior_scale=0.5,
    interval_width=0.95
)
prophet_model.fit(prophet_data)

# Make future dataframe for next year
future = prophet_model.make_future_dataframe(periods=1, freq='Y')

# Predict
forecast = prophet_model.predict(future)

# Get next year's prediction
next_year_forecast = forecast.iloc[-1]

print("FACEBOOK PROPHET FORECAST")
print("="*60)
print(f"Next Year ({int(next_year)}): {int(next_year_forecast['yhat']):,} registrations")
print(f"Lower Bound (95% CI): {int(next_year_forecast['yhat_lower']):,}")
print(f"Upper Bound (95% CI): {int(next_year_forecast['yhat_upper']):,}")
```

```python
# Plot forecast
fig1 = prophet_model.plot(forecast, figsize=(12, 6))
plt.title('Employment Program Registrations - Prophet Forecast', fontsize=16, fontweight='bold')
plt.xlabel('Year', fontsize=12)
plt.ylabel('Registrations', fontsize=12)
plt.tight_layout()
plt.show()

# Plot components
fig2 = prophet_model.plot_components(forecast)
plt.tight_layout()
plt.show()
```

```python
# Summary comparison
comparison = pd.DataFrame({
    'Method': ['Linear Regression', 'ARIMA', 'Prophet'],
    'Forecast': [
        int(next_year_pred),
        int(best_model.forecast(steps=1).iloc[0]),
        int(next_year_forecast['yhat'])
    ]
})

print("FORECAST COMPARISON FOR", int(next_year))
print("="*60)
print(comparison.to_string(index=False))

# Visualize comparison
fig, ax = plt.subplots(figsize=(12, 6))

# Plot historical data
ax.plot(yearly_data['start_year'], yearly_data['registrations'], 
        marker='o', linewidth=2, markersize=8, label='Historical', color='black')

# Plot forecasts
ax.scatter([next_year], [next_year_pred], s=200, marker='s', 
           label=f'Linear Regression: {int(next_year_pred):,}', color='blue', zorder=5)
ax.scatter([next_year], [int(best_model.forecast(steps=1).iloc[0])], s=200, marker='^', 
           label=f'ARIMA: {int(best_model.forecast(steps=1).iloc[0]):,}', color='red', zorder=5)
ax.scatter([next_year], [int(next_year_forecast['yhat'])], s=200, marker='D', 
           label=f'Prophet: {int(next_year_forecast["yhat"]):,}', color='green', zorder=5)

ax.set_xlabel('Year', fontsize=12)
ax.set_ylabel('Registrations', fontsize=12)
ax.set_title('Employment Program Registrations - Forecast Comparison', fontsize=16, fontweight='bold')
ax.legend(loc='best', fontsize=10)
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```
