```python
import glob
import pandas as pd

# Get a list of all CSV files that match the pattern
csv_files = glob.glob('/content/daily-shelter-overnight-service-occupancy-capacity-*.csv')

# Initialize an empty list to store individual DataFrames
dfs = []

# Loop through each file, read it, and append to the list
for file in csv_files:
    df = pd.read_csv(file)
    df.info()
    df['OCCUPANCY_DATE'] = pd.to_datetime(
    df['OCCUPANCY_DATE'].astype(str),
    yearfirst=True,   # treat the first number as the year
    errors='coerce'
)
    dfs.append(df)

# Concatenate all DataFrames into a single DataFrame
merged_df = pd.concat(dfs, ignore_index=True)
merged_df = merged_df.sort_values('OCCUPANCY_DATE')
```

### Dropping Data-Heavy Missing Columns

Columns with more than 60% missing values will be dropped. These include `CAPACITY_ACTUAL_ROOM`, `CAPACITY_FUNDING_ROOM`, `OCCUPIED_ROOMS`, `UNOCCUPIED_ROOMS`, `UNAVAILABLE_ROOMS`, and `OCCUPANCY_RATE_ROOMS`.

```python
columns_to_drop = missing_percentages[missing_percentages > 60].index.tolist()
merged_df = merged_df.drop(columns=columns_to_drop)

print(f"Dropped columns: {columns_to_drop}")
print("Updated merged_df info after dropping columns:")
print(merged_df.info())
```

### Calculate Outlier Boundaries and Remove Outliers (IQR Method)

This code calculates the Interquartile Range (IQR) for each numerical column in `merged_df` and defines outlier bounds. It then filters `merged_df` to remove any rows where a numerical value falls outside these bounds, creating a new `merged_df_cleaned` DataFrame.

```python
import pandas as pd
import numpy as np

# Identify numerical columns in the current merged_df
num_cols_for_outliers = merged_df.select_dtypes(include=['number']).columns.tolist()

# Exclude '_id' column if present, as it's typically an identifier, not a measurement
if '_id' in num_cols_for_outliers:
    num_cols_for_outliers.remove('_id')

# Initialize a combined mask to keep all rows initially
combined_mask = pd.Series(True, index=merged_df.index)

print("Calculating Outlier Boundaries (IQR Method) and Identifying Outliers:")
print("---------------------------------------------------")

for col in num_cols_for_outliers:
    # Calculate Q1 (25th percentile) and Q3 (75th percentile)
    Q1 = merged_df[col].quantile(0.25)
    Q3 = merged_df[col].quantile(0.75)
    IQR = Q3 - Q1

    # Calculate outlier bounds
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    print(f"\nColumn: {col}")
    print(f"  Q1: {Q1:.2f}, Q3: {Q3:.2f}, IQR: {IQR:.2f}")
    print(f"  Lower Bound: {lower_bound:.2f}")
    print(f"  Upper Bound: {upper_bound:.2f}")

    # Create a mask for the current column: True if within bounds, False if outlier
    # Handle NaN values by making their condition False to drop them if they exist in relevant columns
    col_mask = (merged_df[col] >= lower_bound) & (merged_df[col] <= upper_bound)
    col_mask = col_mask.fillna(False) # Treat NaNs as outside the bounds for filtering purposes

    # Combine this mask with the overall mask using logical AND
    # A row will be kept only if it's NOT an outlier in ANY of the numerical columns
    combined_mask = combined_mask & col_mask
    print(f"  Rows identified as outliers in '{col}': {(~col_mask).sum()}")

print("---------------------------------------------------")

# Apply the combined mask to the DataFrame to remove rows with any identified outliers
merged_df_cleaned = merged_df[combined_mask].copy()

print(f"\nOriginal number of rows: {len(merged_df)}")
print(f"Number of rows after removing outliers from all numerical columns: {len(merged_df_cleaned)}")
print(f"Total outliers removed: {len(merged_df) - len(merged_df_cleaned)}\n")

display(merged_df_cleaned.head())
print(merged_df_cleaned.info())
```

### Imputing Missing Categorical Data with Mode

```python
cols_to_impute_mode = [
    'SHELTER_GROUP',
    'LOCATION_NAME',
    'LOCATION_ADDRESS',
    'LOCATION_POSTAL_CODE',
    'LOCATION_CITY',
    'LOCATION_PROVINCE',
    'PROGRAM_MODEL'
]

for col in cols_to_impute_mode:
    if merged_df_cleaned[col].isnull().any():
        mode_value = merged_df_cleaned[col].mode()[0] # [0] to handle cases where there are multiple modes
        merged_df_cleaned[col].fillna(mode_value, inplace=True)
        print(f"Column '{col}' imputed with mode: '{mode_value}'")

print("\nMissing values after imputation:")
print(merged_df_cleaned.isnull().sum())

print("\nUpdated merged_df_cleaned info after imputation:")
print(merged_df_cleaned.info())
```

### Recalculate Outlier Boundaries and Remove Outliers from the Updated Data (IQR Method)


```python
import pandas as pd
import numpy as np

# Identify numerical columns in the current merged_df_cleaned
num_cols_for_outliers_v2 = merged_df_cleaned.select_dtypes(include=['number']).columns.tolist()

# Exclude '_id' column if present, as it's typically an identifier, not a measurement
if '_id' in num_cols_for_outliers_v2:
    num_cols_for_outliers_v2.remove('_id')

# Initialize a combined mask to keep all rows initially
combined_mask_v2 = pd.Series(True, index=merged_df_cleaned.index)

print("Recalculating Outlier Boundaries (IQR Method) and Identifying Outliers on Updated Data:")
print("---------------------------------------------------")

for col in num_cols_for_outliers_v2:
    # Calculate Q1 (25th percentile) and Q3 (75th percentile)
    Q1 = merged_df_cleaned[col].quantile(0.25)
    Q3 = merged_df_cleaned[col].quantile(0.75)
    IQR = Q3 - Q1

    # Calculate outlier bounds
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    print(f"\nColumn: {col}")
    print(f"  Q1: {Q1:.2f}, Q3: {Q3:.2f}, IQR: {IQR:.2f}")
    print(f"  Lower Bound: {lower_bound:.2f}")
    print(f"  Upper Bound: {upper_bound:.2f}")

    # Create a mask for the current column: True if within bounds, False if outlier
    # Handle NaN values by making their condition False to drop them if they exist in relevant columns
    col_mask = (merged_df_cleaned[col] >= lower_bound) & (merged_df_cleaned[col] <= upper_bound)
    col_mask = col_mask.fillna(False) # Treat NaNs as outside the bounds for filtering purposes

    # Combine this mask with the overall mask using logical AND
    # A row will be kept only if it's NOT an outlier in ANY of the numerical columns
    combined_mask_v2 = combined_mask_v2 & col_mask
    print(f"  Rows identified as outliers in '{col}': {(~col_mask).sum()}")

print("---------------------------------------------------")

# Apply the combined mask to the DataFrame to remove rows with any identified outliers
merged_df_cleaned = merged_df_cleaned[combined_mask_v2].copy()

print(f"\nOriginal number of rows (after previous cleaning): {len(combined_mask_v2)}")
print(f"Number of rows after removing additional outliers from all numerical columns: {len(merged_df_cleaned)}")
print(f"Total additional outliers removed: {len(combined_mask_v2) - len(merged_df_cleaned)}\n")

display(merged_df_cleaned.head())
print(merged_df_cleaned.info())
```
